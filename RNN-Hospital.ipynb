```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Model, load_model
from keras import Input
from keras.layers import Dense, SimpleRNN
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/Hospital_63-64.csv')
```

```python
df
```

```python
X = df[['ม.ค.', 'ก.พ.', 'มี.ค.', 'เม.ย.', 'พ.ค.', 'มิ.ย.', 'ก.ค.', 'ส.ค.', 'ก.ย.', 'ต.ค.', 'พ.ย.', 'ธ.ค.']]
```

```python
X.shape
```

```python
y = df[['ม.ค..1', 'ก.พ..1', 'มี.ค..1', 'เม.ย..1', 'พ.ค..1', 'มิ.ย..1', 'ก.ค..1', 'ส.ค..1', 'ก.ย..1', 'ต.ค..1', 'พ.ย..1', 'ธ.ค..1']]
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_train_transformed = (X_train.to_numpy()).reshape(len(X_train.to_numpy()), 12, 1)
```

```python
X_train_transformed.shape
```

```python
X_test_transformed = (X_test.to_numpy()).reshape(len(X_test.to_numpy()), 12, 1)
```

```python
X_test_transformed.shape
```

```python
y_train_transformed = (y_train.to_numpy()).reshape(len(y_train.to_numpy()), 12, 1)
```

```python
y_train_transformed.shape
```

```python
y_test_transformed = (y_test.to_numpy()).reshape(len(y_test.to_numpy()), 12, 1)
```

```python
y_test_transformed.shape
```

```python
inputLayer = Input(shape=(12, 1))
hidden1 = SimpleRNN(8, activation='relu', return_sequences=True, dropout=0.1)(inputLayer)
outputLayer = Dense(1, activation='relu')(hidden1)
model = Model(inputs=inputLayer, outputs=outputLayer)

model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['MeanAbsoluteError'])

model.summary()
```

```python
history = model.fit(X_train_transformed, y_train_transformed, batch_size=3, epochs=50, validation_split=0.2)
```

```python
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
pred_train = model.predict(X_train_transformed)
```

```python
pred_train.shape
```

```python
pred_train_transformed = pred_train.reshape(len(pred_train), 12)
```
